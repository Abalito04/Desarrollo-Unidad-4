```python

import os
import pandas as pd
import matplotlib.pyplot as plt
# 1. Asegurarnos de estar en el repo
%cd /content/Desarrollo-Unidad-4
# 2. (P1) Crear datos de ejemplo en /datos (Escenario B)
os.makedirs('datos', exist_ok=True)
data = {
    'sales_date': ['2024-01-01', '2024-01-02', '2024-02-01', '2024-02-15', '2024-03-01'],
    'product': ['Producto A', 'Producto B', 'Producto A', 'Producto C', 'Producto B'],
    'quantity': [10, 5, 8, 12, 7],
    'amount': [1500, 2500, 1200, 3000, 3500]
}
df = pd.DataFrame(data)
df.to_csv('datos/ventas.csv', index=False)
# 3. (P2) Crear el SCRIPT REAL de análisis en /scripts
os.makedirs('scripts', exist_ok=True)
with open('scripts/analisis_ventas.py', 'w') as f:
    f.write("""
import pandas as pd
import matplotlib.pyplot as plt
# Cargar datos
df = pd.read_csv('datos/ventas.csv')
df['sales_date'] = pd.to_datetime(df['sales_date'])
# Calcular indicadores
ventas_totales = df['amount'].sum()
producto_mas_vendido = df.groupby('product')['quantity'].sum().idxmax()
print(f'Ventas Totales: ${ventas_totales}')
print(f'Producto más vendido: {producto_mas_vendido}')
# Generar gráfico
df.groupby('sales_date')['amount'].sum().plot(kind='line', marker='o')
plt.title('Evolución de Ventas 2024')
plt.xlabel('Fecha')
plt.ylabel('Monto ($)')
plt.savefig('resultados/grafico_ventas.png')
""")
# 4. (P2) Ejecutar el script para generar los resultados
os.makedirs('resultados', exist_ok=True)
!python scripts/analisis_ventas.py
# 5. (P3) Actualizar el README con la info del escenario
with open('README.md', 'w') as f:
    f.write('# Análisis de Ventas - Empresa Simulada (Escenario B)\n\n')
    f.write('**Integrante:** Abalito04\n')
    f.write('**Descripción:** Este proyecto analiza el desempeño comercial de una PyME.\n')
    f.write('**Dataset:** Ventas simuladas 2024.\n')
    f.write('**Instrucciones:** Ejecutar scripts/analisis_ventas.py para regenerar resultados.\n')
# 6. Subir todo a GitHub con trazabilidad
!git add .
!git commit -m "DU4-2: feat: implementacion de Escenario B con analisis de ventas y graficos"
!git push https://{TOKEN}@github.com/Abalito04/Desarrollo-Unidad-4.git main
```
